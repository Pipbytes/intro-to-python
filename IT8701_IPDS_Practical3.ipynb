# 3. Working with multiple figures and axes

### Task 1: Create a multi-plot

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
# Initialize chart 1 array
y1 = np.arange(1,10,0.5)
y1_color = 'b'
y1_marker = '*'

# Initialize chart 2
y2 = np.arange(1,50,5)
y2_color = '#ff69b4'
y2_marker = 's'

# Initialize chart 3
y3 = np.arange(100,200,10)
y3_color = 'g'
y3_marker = 'd'

# Initialize chart 4
y4 = np.arange(1000,2000,50)
y4_color = ((1.0, 0.5, 0.5)) #rgb
y4_style = '-'

# # Create a figure and a grid of subplots with 4 rows and 1 column
fig, axes = plt.subplots(4, 1, figsize=(10,6))
plt.figure(1)

# Subplot 1a
axes[0].plot(y1, color=y1_color, marker=y1_marker, linestyle='None')

# Subplot 1b
axes[1].plot(y2, color=y2_color, marker=y2_marker, linestyle='None')

# Subplot 1c
axes[2].plot(y3, color=y3_color, marker=y3_marker, linestyle='None')

# Subplot 1d
axes[3].plot(y4, color=y4_color, linestyle=y4_marker)


# Create a figure and a grid of subplots with 2 rows and 2 column
plt.figure(2)

# Subplot 2a
plt.subplot(221) # 2 rows, 2 cols, chart1
plt.plot(y1, color=y1_color, marker=y1_marker, linestyle='None')

# Subplot 2b
plt.subplot(222) # 2 rows, 2 cols, chart2
plt.plot(y2, color=y2_color, marker=y2_marker, linestyle='None')

# Subplot 2c
plt.subplot(223) # 2 rows, 2 cols, chart3
plt.plot(y3, color=y3_color, marker=y3_marker, linestyle='None')

# Subplot dd
plt.subplot(224) # 2 rows, 2 cols, chart4
plt.plot(y4, color=y4_color, linestyle=y4_style)

# Adjust layout
plt.tight_layout()

plt.show()
```

# 4. Working with Text

### Task 1: Add text and "text-in-box"

```python
# Import library
import matplotlib.pyplot as plt
import numpy as np

# Initialize array
x = np.arange(0.0, 5.0, 0.01)
y = np.sin(2 * np.pi * x)

# Plot the figure
plt.plot(x, y, color='r')

# Annotate graph with text
plt.text(2,1, 'x=2, y=1')
plt.text(5,1, 'x=5, y=1', bbox=dict(facecolor='red', alpha=0.5), verticalalignment='center')

# Set limit and labels
plt.xlim(0, 6)
plt.ylim(-1.5, 1.5)

# Show the plot
plt.show()
```

### Task 2: Add annotation (arrow)

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt

# Initialize array
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2 * np.pi * x)

# Create the plot
plt.plot(x, y, 'b-')

# Annotate the graph with arrows and text
plt.annotate('x=2, y=1', xy=(2, 1), xytext=(0.5, 1.5), arrowprops=dict(facecolor='blue', shrink=0.05))
plt.annotate('x=1.5,y=-1', xy=(1.5, -1), xytext=(0.5, -1.5), arrowprops=dict(facecolor='red', shrink=0.05))

# Set the limits and labels
plt.xlim(0, 5)
plt.ylim(-2, 2)

# Show the plot
plt.show()
```

# 5. Bar Charts

### Task 1: Analyze live births

```python
# import library
import numpy as np
import matplotlib.pyplot as plt

# Initialize dataset
fname = 'live-births.csv'
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('year', 'i8'),('level_1','U50'),('value','i8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0])

# Filter the data for 'Total Live-Births' and year 2006 to 2016
filtered_data = data[(data['level_1'] == 'Total Live-Births') & (data['year'] >= 2006) & (data['year'] <= 2016)]

# function to add value labels
def addlabels(x,y):
  for i in range(len(x)):
    plt.text(x[i], y[i]//2, y[i], ha = 'center')

# Plot Vertical Bar Chart
plt.figure(figsize=(10,6))
plt.bar(filtered_data['year'], filtered_data['value'], color='red')
plt.ylabel('Total Live Births')
plt.title('Live Births over the years', fontsize=18)
plt.xticks(filtered_data['year'], rotation=90)
addlabels(filtered_data['year'], filtered_data['value'])


# Show plot
plt.show()
```

# 6. Pie Charts

### Task 1: Analyze gender comparison in years 1960 vs 2016

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt

# Define the file path
fname = 'singapore-residents-by-ethnic-group-and-sex-end-june-annual.csv'

# Load the CSV file using numpy
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('year','i8'),('level_1','U50'),('value','i8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0])

# Extract year 1960 & 2016, and gender composition of total male & female resident
gender_comp1960 = data[(data['year'] == 1960) & (np.isin(data['level_1'], ['Total Male Residents', 'Total Female Residents']))]
gender_comp2016 = data[(data['year'] == 2016) & (np.isin(data['level_1'], ['Total Male Residents', 'Total Female Residents']))]

# Subplot pie chart 1960
plt.figure(1, figsize=(8,4))
plt.subplot(121) # 1 row, 2 cols, chart1
plt.pie(gender_comp1960['value'], labels=gender_comp1960['level_1'], autopct='%1.1f%%')
plt.title("Gender Composition in 1960")

# Subplot pie chart 2016
plt.subplot(122) # 1 rows, 2 cols, chart2
plt.pie(gender_comp2016['value'], labels=gender_comp2016['level_1'], autopct='%1.1f%%')
plt.title("Gender Composition in 2016")

# Show plot
plt.show()
```

# 7. Scatter Plots

### Task 1: Analyze the relation between rainfall and temperature with scatterplot

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt

# Initialize data
fname1 = 'surface-air-temperature-monthly-mean.csv'
fname2 = 'rainfall-monthly-number-of-rain-days.csv'
data1 = np.genfromtxt(fname1, skip_header=1,
                      dtype=[('month','datetime64[M]'),('mean_temp','f8')],
                      delimiter=',',
                      missing_values=['na','-'], filling_values=[0])
data2 = np.genfromtxt(fname2, skip_header=1,
                      dtype=[('month','datetime64[M]'),('no_of_rainy_days','i8')],
                      delimiter=',',
                      missing_values=['na','-'], filling_values=[0])

# obtain m (slope) and b (intercept) of linear regression line
m, b = np.polyfit(data1['mean_temp'], data2['no_of_rainy_days'], 1)

# Plot Scatter Plot
plt.scatter(data1['mean_temp'], data2['no_of_rainy_days'],
            c='g')
plt.plot(data1['mean_temp'], m*data1['mean_temp']+b, color='r')
plt.xlabel('Mean Monthly Temperature')
plt.ylabel('Number of Rainy Days')
plt.title('Monthly Rainfall and Temperatures Since Jan 1980')
plt.show()
```

# 9. Box Plots

### Task 1: Analyze median results in different areas using box plot

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize data with numpy
fname = 'median-rentals-and-vacancy-of-retail-space-by-locality-quarterly.csv'
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('quarter','U50'),('locality','U50'),('ret_med_rent_lease_cm',float),('ret_med_rent_cont_date','<f8'),('ret_vacancy_rate','f8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0])

# Replace text 'central area - outside orchard' with 'central area'
locality = np.array([loc.replace('Central Area - Outside Orchard', 'Central Area') for loc in data['locality']])
psf = data['ret_med_rent_lease_cm']

# Prepare data for box plot
localities = np.unique(locality)
boxplot_data = [psf[locality == loc] for loc in localities]

# Calculate median rent for each locality and sort by these values
median_rents = [np.median(psf) for psf in boxplot_data]
sorted_indices = np.argsort(median_rents)[::-1]  # Sort indices by median rent in descending order

# Sort localities and boxplot data according to the sorted indices
sorted_localities = localities[sorted_indices]
sorted_boxplot_data = [boxplot_data[i] for i in sorted_indices]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(sorted_boxplot_data, labels=sorted_localities, boxprops=dict(color='b'), flierprops=dict(markerfacecolor='g'), medianprops=dict(color='r'))
plt.title('Median Rentals in Orchard, Outside Central Area and Central Area')
plt.xlabel('Locality', fontsize=12)
plt.ylabel('$psf')

# Show plot
plt.show()
```

# 10. Interactive Charts

### Task 1: Create interactive chart showing x,y values as textbox

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2 * np.pi * x)

# Create the plot
fig, ax = plt.subplots()
line, = ax.plot(x, y, lw=2)

# Define the callback function to show x and y values on click
def onclick(event):
    if event.inaxes:
        ax.annotate(f'x={event.xdata:.2f}\ny={event.ydata:.2f}',
                    xy=(event.xdata, event.ydata),
                    xytext=(event.xdata+0.5, event.ydata),
                    arrowprops=dict(facecolor='black', shrink=0.05),
                    bbox=dict(facecolor='red', alpha=0.5))
        fig.canvas.draw()

# Connect the click event to the callback function
fig.canvas.mpl_connect('button_press_event', onclick)

# Show plot
plt.show()
```
