```python
# Import library
import numpy as np

# print headers
title = "Number of Hospital Beds"
title_len = len(title)
print(f"{title_len*'*'} {title} {title_len*'*'}\n")

# Initialize data array
fname = "hospital-admissions-and-public-sector-outpatient-attendances-annual.csv"
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('year','U10'),('level_1','U50'),('value','f8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=0) #U10/U50 = length for string, f8=float 8 decimals

# Display array shape
print(f"There are {len(data)} rows in this dataset")

# Display unique valyes in year and level_1 column
set_year = set(data['year'])
set_level_1 = set(data['level_1'])
print(f"There are {len(set_year)} unique values in year column.")
print(f"There are {len(np.unique(data['year']))} unique values in year column.")
print(f"There are {len(set_level_1)} unique values in level_1 column.")
print(f"There are {len(np.unique(data['level_1']))} unique values in level_1 column.\n")

# print the mean number of hospital bed over the years.
print(f"The mean number of hospital bed over the years is {np.mean(data['value']):.0f}")

# print the year with the maximum number of beds
max_value = np.max(data['value'])
max_row_index = np.argmax(data['value'])
print(f"The maximum number of hospital beds in the year was {data['year'][max_row_index]} for Polyclinic with {max_value:.0f} beds.")

# print the year the the minimum number of beds
d = data['value'].copy() # duplicate a data set to replace 0
replace_value = np.max(d)+1 # temp replacement value
d[d==0]=replace_value # replace 0 w temp value
min_value = np.min(d) # find lowest value excluding 0
min_row_index = np.argmin(d) # find row index of lowest value excluding 0
print(f"The minimum number of hospital beds in the year was {data['year'][min_row_index]} for Polyclinic with {min_value:.0f} beds.")

# print the standard deviation
std = np.std(data['value'])
print(f"The standard deviation of the number of hospital beds is {std:.0f}.")
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize data array
fname = "number-of-hospital-beds-annual.csv"
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('year','datetime64[Y]'),('sector','U50'),('no_of_hospital_beds','i8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0]) #U10/U50 = length for string, i8=integer 8 digits


# Filter data based Public Sector + Year and Public Sector + # of hospital beds
x_public = data[data['sector']=='Public Sector']['year']
y_public = data[data['sector']=='Public Sector']['no_of_hospital_beds']

# Filter data based Private Sector + Year and Private Sector + # of hospital beds
x_private = data[data['sector']=='Private Sector']['year']
y_private = data[data['sector']=='Private Sector']['no_of_hospital_beds']

# Plot line chart
fig = plt.figure(figsize=(16,6))
ax1 = fig.add_subplot(111)

ax1.plot(x_public, y_public, label='Public Sector')
ax1.plot(x_private, y_private, label='Private Sector')
plt.ylabel("Number of Hospital Beds")
plt.xticks(rotation=45)
plt.legend(loc='upper left')

plt.show()
```

# Example 4: Create a Barchart

### Method 1

```python
# Import library
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
from operator import itemgetter

# Initialize data array
fname = "hospital-admissions-and-public-sector-outpatient-attendances-annual.csv"
data = np.genfromtxt(fname, skip_header=1,
                     dtype=[('year','U10'),('level_1','U50'),('value','i8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0]) #U10/U50 = length for string, i8=integer 8 digits

# Check original data
print(f"Original data: {data.shape}")

# Check for null rows
null_rows = np.isnan(data['value']) # Return True if row is null
non_null_rows = data[null_rows == False] # Return non-null
print(f"Filtered Data: {non_null_rows.shape}")

# Re-package array
labels = list(set(data['level_1'])) # find unique labels
levels = np.arange(0, len(labels)) # numbers of unique labels
levels_values = data[['level_1','value']] # repackage labels w values, removed year
values = levels_values['value'] # get values

# Compute average values for each labels
avg_values = {} # empty dict
for i in labels:
  values_for_level = values[levels_values['level_1']==i] # extract all values for each labels
  avg = np.average(values_for_level) # find the average of each label using np.average
  print(f"Average value for level {i} is {avg:.0f}.")
  avg_values[i] = avg # store labels in keys() and values as values in values()

# OrderedDict allows barcharts to be displayed in descending order
avg_values = OrderedDict(sorted(avg_values.items(), key=itemgetter(1), reverse=True))

# Plot bar charts
plt.figure(1, figsize=(12,8))
barchart = plt.bar(list(avg_values.keys()), list(avg_values.values()), color='#d62728') # sort dictionary

#
for k in range(len(barchart)):
    bar = barchart[k]
    x,y  = bar.get_xy()
    h = bar.get_height()
    plt.text(x,h,"{:.0f}".format(list(avg_values.values())[k]),fontsize=12)

plt.title('Hospital Admissions and Public Sector Outpatient Attendances Annual',fontsize=18)
plt.ylabel('No. of patients', fontsize=8)
plt.yticks(fontsize=8)
plt.xticks(levels, labels, fontsize=8, rotation=45)
plt.show()
```

# Example 4: Create a Boxplot

### Method 1

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Initialize data array
fname = "hospital-admissions-and-public-sector-outpatient-attendances-annual.csv"
fname2 = "hospital-admissions-and-public-sector-outpatient-attendances-annual-v2.csv"
data = np.genfromtxt(fname2, skip_header=1,
                     dtype=[('year','U10'),('level_1','U50'),('value','i8')],
                     delimiter=',',
                     missing_values=['na','-'], filling_values=[0]) #U10/U50 = length for string, i8=integer 8 digits

# Check original data
print(f"Original data: {data.shape}")

# Check for null rows / Handling Missing Data
null_rows = np.isnan(data['value']) # Return True if row is null
non_null_rows = data[null_rows == False] # Return non-null
print(f"Filtered Data: {non_null_rows.shape}")

# Extracting labels and Data
labels = list(set(data['level_1'])) # find unique labels
levels = np.arange(0, len(labels)) # numbers of unique labels
levels_values = data[['level_1','value']] # repackage labels w values, removed year
values = levels_values['value'] # get values

# Filtering Values for Each Category
values_combined = {}
for label in labels:
    values_combined[label] = data['value'][data['level_1'] == label]

## Creating Box Plot
plt.figure(2, figsize=(12,8))
plt.title('Hospital Admission',fontsize=8)
plt.ylabel('Number of patients',fontsize=8)
plt.yticks(fontsize=8)
plt.xticks(fontsize=8,rotation='vertical')

# v_labels = ['AE', 'Day Surgeries', 'Dental Clinics', 'Polyclinics',
#           'Specialist Outpatient Clinics', 'Total Hospital Admissions']

bp_dict = plt.boxplot(values_combined.values(),labels=labels,patch_artist=True)

## Customize box plot
# change outline color, fill color and linewidth of the boxes
for box in bp_dict['boxes']:
  box.set(color='#7570b3', linewidth=2)
  box.set(facecolor = '#1b9e77')

# change color and linewidth of the whiskers
for whisker in bp_dict['whiskers']:
   whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp_dict['caps']:
   cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp_dict['medians']:
   median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp_dict['fliers']:
   flier.set(marker='D', color='#e7298a', alpha=0.5)

print(bp_dict.keys())

## Adding Annotations
for line in bp_dict['medians']:
  # get position data for median line
  x, y = line.get_xydata()[1] # top of median line
  # overlay median value
  plt.text(x, y, '%.1f' % y, horizontalalignment='center',fontsize=8) # draw above, centered

fliers = []
for line in bp_dict['fliers']:
  ndarray = line.get_xydata()
  if (len(ndarray)>0):
    max_flier = ndarray[:,1].max()
    max_flier_index = ndarray[:,1].argmax()
    x = ndarray[max_flier_index,0]
    print("Flier: " + str(x) + "," + str(max_flier))
    plt.text(x,max_flier,'%.1f' % max_flier,horizontalalignment='center',fontsize=8,color='green')

plt.show()
```
