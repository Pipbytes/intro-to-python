### Select columns by Regular Expression (RE)

```python
import pandas as pd
import numpy as np

# Initialize data
fname = 'singstats_maritalstatus.xlsx'
mydf = pd.read_excel(fname, index_col='Variables')

# Establish RE search parameter
re1 = '\.' # Match strings containing a period
re2 = '201$' # Match strings ending with p
re3 = '^c' # Match strings starting with c
re4 = '^c[1-5]$' # Match strings starting with c and ending with 1,2,3,4,5
re5 = '^(?!pop$).*' # Match strings except those containing pop
re6 = '^201' # Match strings starting with ...

# Filter columns that starts with 2010 onwards
df_2010_and_after = mydf.filter(regex=re6)
print(f"*** Columns starting with 2010 ***\n{df_2010_and_after.to_string(index=True)}\n")
```

## 7.1 Drop columns/rows from DataFrame

### Drop columns and rows from a DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Initialize data
fname = 'graduate-employment-survey-ntu-nus-sit-smu-sutd.csv'
df = pd.read_csv(fname, encoding='ISO-8859-1')


# Operation 1 - Remove the following columns in df
removal = ['school', 'employment_rate_overall', 'employment_rate_permanent', 'basic_monthly_mean',
           'basic_monthly_median', 'gross_mthly_25_percentile', 'gross_mthly_75_percentile']
df2 = df.copy()
df2 = df2.drop(columns=removal)


# Operation 2 - Remove the following names in university column
uni = ['Nanyang Technological University','Singapore Institute of Technology','Singapore University of Technology and Design']
df3 = df2.copy()
index_to_drop = df3[df3['university'].isin(uni)].index
df3 = df3.drop(index_to_drop, axis=0)
#df3 = df3[~df3['university'].isin(uni)] # alternate solution not using drop() method


# Operation 3 - Remove data that contain 2013 and 2014 in the 'year' column
period = [2013, 2014]
df4 = df3.copy()
year_to_drop = df4[df4['year'].isin(period)].index
df4 = df4.drop(year_to_drop, axis=0)
#df4 = df4[~df4['year'].isin(period)


# Operation 4 - Remove data that is less than 4000 in the 'gross_monthly_median' column
df5 = df4.copy()
df5['gross_monthly_median'] = pd.to_numeric(df5['gross_monthly_median'], errors='coerce')
df5 = df5[df5['gross_monthly_median'] >= 4000]
df5.drop(columns=['year','university'], inplace=True)


print(f"Shape of df in its original form = {df.shape}")
print(f"Shape of df after dropping unwanted columnns = {df2.shape}")
print(f"Shape of df after dropping unwanted rows (university) = {df3.shape}")
print(f"Shape of df after dropping unwanted rows (year) = {df4.shape}")
print(f"Shape of df after dropping unwanted rows (gross_monthly_median) = {df5.shape}")
print("")
print("*** Final Data after ***")
print(df5)
```

## 8.3 Fill missing values (fillna)

### Fill missing values with a scalar value

```python
import pandas as pd
import numpy as np

# Load data
mydf = pd.read_csv('hdbresaleindex_missing.csv', index_col='quarter')

# Use the isnull method to identify columns with missing or invalid data
df_missing = mydf[mydf.isnull().any(axis=1)]

# Print data
print("***** Rows with missing values -- df_missing *****")
print(df_missing)
print("")

# Fill missing values with scalar value 50.0
df_filled = mydf.fillna(50.0)
print("***** Dataset after scalar fill operation")
print(df_filled.head(10).to_string(index=True))
```

### Fill missing gaps forwards

```python
import pandas as pd
import numpy as np

# Load data
mydf = pd.read_csv('hdbresaleindex_missing.csv', index_col='quarter')

# Use the isnull method to identify columns with missing or invalid data
df_missing = mydf[mydf.isnull().any(axis=1)]

# Print data
print("***** Rows with missing values -- df_missing *****")
print(df_missing)
print("")

# Fill missing values with forward fill operation
df_filled = mydf.ffill(axis=0)
print("***** Dataset after Forward fill operation")
print(df_filled.head(15).to_string(index=True))
```

### Filling missing gaps backwrds

```python
import pandas as pd
import numpy as np

# Load data
mydf = pd.read_csv('hdbresaleindex_missing.csv', index_col='quarter')

# Use the isnull method to identify columns with missing or invalid data
df_missing = mydf[mydf.isnull().any(axis=1)]

# Print data
print("***** Rows with missing values -- df_missing *****")
print(df_missing)
print("")

# Fill missing values with backward fill operation
df_filled = mydf.bfill(axis=0)
print("***** Dataset after Backward Fill operation")
print(df_filled.head(15).to_string(index=True))
```

### Filling missing gaps with mean of column

```python
import pandas as pd
import numpy as np

# Load data
mydf = pd.read_csv('hdbresaleindex_missing.csv', index_col='quarter')

# Rename column
# mydf.rename(columns={'index':'resale_index'}, inplace=True)
mydf.columns = ['resale_index']

# Print data
print("***** First 5 rows of original data set")
print(mydf.head())
print("")
print("***** Last 5 rows of original data set")
print(mydf.tail())
print("")

# Use the isnull method to identify columns with missing or invalid data
df_missing = mydf[mydf.isnull().any(axis=1)]

# Print data
print("***** Rows with missing values -- df_missing *****")
print(df_missing)
print("")

# Fill missing values with mean of column
df_filled = mydf.fillna(mydf.mean())
print(f"Mean of resale index is {mydf.mean()}")
print("")
print("***** Dataset after mean fill operation")
print(df_filled.head(15))
```

### Combine two datasets with merge method

```python
import pandas as pd

# Load data
df_m = pd.read_csv('merge_students-male.csv')
df_f = pd.read_csv('merge_students-female.csv')
df_test = pd.read_csv('merge_testscores.csv')

# Combine df_female and df_male into a single dataset df_students using concat
df_students = pd.concat([df_f, df_m])

# Merge df_students and df_test using the studentid value (INNER JOIN) into a dataset df_merged
df_merged = pd.merge(df_students, df_test, on='studentid', how='inner')

# Display the merged DataFrame
print(f"Total students: {len(df_merged)}")
print("")
print("Test scores in descending order")
print(df_merged.sort_values(by=['test score','studentid'], ascending=[False,True]).head(10))
```
