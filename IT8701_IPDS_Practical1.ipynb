### Task 3: Days before assignment is due

```python
from datetime import datetime
from dateutil import relativedelta as rdelta
```

```python
# Today's date in datetime format
date_today = datetime.now()
print(f"Today is {date_today:%d-%b-%y}")

# Prompt user with a message
assignment_duedate = input("Please enter the due date of the assignment (dd mmm yy): ")

# Convert string to Datetime format
assignment_duedate = datetime.strptime(assignment_duedate, '%d %b %y')

# Calculate difference
rd = rdelta.relativedelta(assignment_duedate, date_today)

print(f"Assignment due date is {assignment_duedate:%d-%b-%Y}")
print(f"{rd.months} months and {rd.days} days till assignment is due.")
```

# 4. Working with Python numeric types

### Task 1: Perform basic mathematical calculations

```python
# Prompt user to input numbers
n1 = int(input("Enter number 1: "))
n2 = int(input("Enter number 2: "))
n3 = int(input("Enter number 3: "))
n4 = int(input("Enter number 4: "))

print(f"The four numbers are: {n1}, {n2}, {n3}, {n4}")

# Sum of all input
add = n1+n2+n3+n4
print(f"Sum is {add}")

# Difference of n1 and n4
difference = n4-n1
print(f"Difference is {difference}")

# Multiplication of n2 and n3
product = n2*n3
print(f"Product is {product}")

# Division the sum of n1 and n2 by sum of n3 and n4
quotient = (n1+n2)/(n3+n4)
print(f"Quotient is {quotient}")

# Remainder of n1 & n2, divided by the product of n3 & n4.
remainder = (n1*n2)%(n3*n4)
print(f"Remainder is {remainder}")

# Exponential of n1 to the power of n2.
power = n1**n2
print(f"Power is {power}")
```

### Task 2: Mathematical functions in Python

```python
# Import library
import math

# Prompt user to input numbers
n1 = float(input("Enter number 1: "))
n2 = float(input("Enter number 2: "))
n3 = float(input("Enter number 3: "))
n4 = float(input("Enter number 4: "))

# Square root of n1, and return the ceiling value
n1_squareroot = math.sqrt(n1)
print(f"The square root of n1_squareroot is {n1_squareroot}.")
print(f"The ceiling value of n1_squareroot is {math.ceil(n1_squareroot)}.\n")

# Multiple n2 by pi
n2_times_pi = n2*math.pi
print(f"{n2} times {math.pi} is {n2_times_pi}.")
print(f"The truncated value of n2_times_pi is {math.trunc(n2_times_pi)}.\n")

# Value of n3 to power of n1, and return the floor value
n3_powerof_n1 = n3**n1
print(f"{n3} to the power of {n1} is {n3_powerof_n1}.")
print(f"The floor value of n3_powerof_n1 is {math.floor(n3_powerof_n1)}.\n")

# Factorial of n4, and divide by n3
n4_factorial = math.factorial(int(n4))
n4n3 = n4_factorial/n3
print(f"{n4} factorial is {n4_factorial}.")
print(f"{n4_factorial} divided by {n3} is {n4n3}.")
print(f"The fractional (remainder) value of the result is {n4_factorial % n3}.")
print(f"The interger (quotient) value of the result is {n4_factorial//n3}.")


```

### Task 2: Spdonalds (SUBMISSION required)

```python
# Write a program that displays a breakfast menu to the user, ask for the food he/she wants to buy and calculate the amount payable
menu = {
    1: {"Item": "SPMuffin", "Price": 5.00},
    2: {"Item": "SPancakes", "Price": 3.00},
    3: {"Item": "SPHashbrown", "Price": 1.50},
}

selection = {}
invalid = True
count = 1

# Display the menu for user to choose
print("Welcome to SPdonalds! \nBelow is our Breakfast menu: ")

while invalid:
  for k,v in menu.items():
    print(f"{k}. {v['Item']} ${v['Price']:,.2f}")

  try:
    # Prompt user to enter the their breakfast choice
    choice = int(input("Enter your choice of food: "))

    # Check if choice is valid:
    if choice not in menu.keys():
      raise ValueError
    # Add to selection
    selection[f'order {count}'] = menu[choice]

    # Prompt user to enter the quantity they want to buy
    quantity = int(input(f"How many {selection[f'order {count}']['Item']}s would you like to buy? "))

    # Check if qty is valid
    if quantity <= 0:
      raise ValueError

    # Add to selection
    selection[f'order {count}']['Quantity'] = quantity
    selection[f'order {count}']['Total'] = quantity * selection[f'order {count}']['Price']

    # Check to continue order
    last_order = input("Anything else? (Y/N)")
    if last_order.upper() == "N":
      total = sum(order['Total'] for order in selection.values())
      print(f"Your total is ${total:.2f}")
      invalid = False
    else: count += 1

  except ValueError:
    print("Sorry, you have entered an invalid choice or quantity. \nUnable to continue. Exiting program...")
```

# 7. for loop

### Task 2: Calculate sum of numbers within a range (SUBMISSION required)

```python
# Inform user to input two numbers from x to y.
print("This program prints the sum of a range of numbers from x to y.")

try:
  # Prompt user to input values of x and y
  x = input("Please enter a value of x: ")
  y = input("Please enter a value of y: ")

  # Check if x and y are numeric
  if not x.isnumeric() or not y.isnumeric():
    raise ValueError("One or more of your inputs are not numeric!")

  # Convert inputs into integer
  x = int(x)
  y = int(y)

  # Check if x and y are higher than zero
  if x<= 0 or y<= 0:
    raise ValueError("One of more of your inputs are not greater than zero.")

  # Check if x < y
  if x>y:
    raise ValueError("You did not enter a value of u that is greaer than x.")

  sum_of_numbers = 0
  for i in range(int(x),int(y)+1):
    sum_of_numbers += i
  print(f"The sum of numbers between {x} and {y} is {sum_of_numbers}")

except ValueError as e:
  print(str(e))
  print("Unable to continue. Exiting Program.. ")
```

# 8. while loop

### Task 2: High-Low number game

```python
# Import library
import random

# Ask user for lower limit and the upper limit of a random number
print("This program will generate a random number between a lower and upper limit of your choice.")
lower = int(input("Enter the lower limit: "))
upper = int(input("Enter the upper limit: "))

# Generate random number
secret_number = random.randint(lower,upper)
print(f"secret num: {secret_number}")

# Write a loop to prompt the user to guess the correct random number
invalid = True
while invalid:
  try:
    # Prompt user to guess the random number
    guess = int(input(f"Enter your guess ({lower} to {upper}): "))

    # Check if guess is within range
    if guess<lower or guess>upper:
      raise ValueError("Your guess is out of range.")

    # Check if guess is above or above the secret number
    if guess>secret_number:
      raise ValueError("Your guess is too high.")
    if guess<secret_number:
      raise ValueError("Your guess is too low.")
    else:
      print("You are right.")
      invalid = False

  except ValueError as e:
    print(str(e))
    print("Please try again.")

```

# 10. Python Functions

### Task 2: Odd and Even (SUBMISSION Required)

```python
import random

# Generate a list of 100 random numbers in the range of 1 to 1000
original_list = [random.randint(1,1000) for x in range(0,100)]

# Define a function that sort
def oddandeven(list):
  odd_list = []
  even_list = []
  for i in list:
    if i%2==0:
      even_list.append(i)
    else:
      odd_list.append(i)
  return odd_list, even_list

odd_numbers, even_numbers = oddandeven(original_list)
print(odd_numbers)
print(even_numbers)
```
